Explain the sentiment analysis techniques used (e.g., TextBlob, NLTK) and provide code snippets.

##**Sentiment Analysis Techniques Used**
**TextBlob**
**Overview:** TextBlob is a simple yet effective library for processing textual data. It provides a straightforward API for common natural language processing (NLP) tasks, including sentiment analysis.

**Sentiment Analysis:** TextBlob uses a predefined set of rules and a lexicon to calculate sentiment scores. Each review is analyzed to produce two scores:

**Polarity:** Ranges from -1 (very negative) to +1 (very positive), indicating the overall sentiment of the review.

**Subjectivity:** Ranges from 0 (objective) to 1 (subjective), measuring how subjective or opinion-based the text is.

**Implementation:** For each review, the TextBlob object is created, and the sentiment method is called to obtain polarity and subjectivity scores.


```python
from textblob import TextBlob
def text_preprocess(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    if(polarity<0 ):
       return 'negative'
    elif(polarity>0):
        return 'positive'
    else:
        return 'neutral'
```

```python
df=df[~df['content'].isnull()]
```

```python
df['sentiment']=[text_preprocess(text) for text in df['content']]
```

Gemini AI is a platform that allows users to analyze and derive insights from large datasets. When using Gemini AI with API keys, the following general steps are involved:

**Authentication:** API keys are used to authenticate your application or script with the Gemini AI platform. This ensures secure access to the API endpoints.

**Data Ingestion:** You can send data to Gemini AI using specific API calls. This can involve uploading datasets, such as reviews from Amazon, for analysis.

**Query Execution**: After the data is ingested, you can use various API endpoints to execute queries. These queries can extract specific insights, such as sentiment analysis, keyword extraction, or trend identification based on the uploaded data.

**Response Handling**: The API returns the results of the analysis in a structured format (often JSON), which can then be processed or visualized for further insights.

**Integration:** The extracted insights can be integrated into applications or dashboards for reporting or further analysis.

Overall, using API keys with Gemini AI facilitates secure and efficient data analysis and insight extraction from large datasets.

```python
import pandas as pd
import google.generativeai as genai

# Load your dataset
df = pd.read_csv('/content/merged_data.csv')

# Replace 'your_api_key' with the actual API key
api_key = "YOUR API KEY"

# Function to analyze reviews using Gemini AI
def analyze_reviews(reviews, api_key,prompt_message):
    genai.configure(api_key=api_key)

    model = genai.GenerativeModel(
        model_name="gemini-1.5-flash",
        generation_config={
            "temperature": 1,
            "top_p": 0.95,
            "top_k": 64,
            "max_output_tokens": 8192,
            "response_mime_type": "text/plain",
        },
    )

    chat_session = model.start_chat(history=[])

    prompt = prompt_message + "\n\n".join(str(review) for review in reviews)
    response = chat_session.send_message(prompt)
    return response.text
```

```python
from textblob import TextBlob
def text_preprocess(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    if(polarity<0 ):
       return 'negative'
    elif(polarity>0):
        return 'positive'
    else:
        return 'neutral'
df=df[~df['content'].isnull()]
df['sentiment'] = [text_preprocess(text) for text in df['content']]


rating_count_per_color = df.groupby(['Company','sentiment']).size().unstack(fill_value=0)
rating_percentage_per_color = rating_count_per_color.div(rating_count_per_color.sum(axis=1), axis=0) * 100
rating_percentage_per_color.plot(kind='bar', stacked=False, figsize=(12, 7))

plt.title('Sentiment Distribution per product (in Percentage)')
plt.xlabel('Company')
plt.ylabel('Percentage of sentiments')
plt.xticks(rotation=45)
plt.legend(title='Sentiments')
plt.tight_layout()

# Display the plot
plt.show()
```
